```python
### look over code and generalise

def createDict(Data, labels):
    """label should be a 2D np array, 2nd order should be a ascending order"""
    Dic = {}
    for i in labels[:, 0]:
         Dic[i] = 0
    
    for val in Data:
        for i in range(len(labels)):
            if val > float(labels[i, 1]):
                pass
            else:
                Dic[labels[i, 0]] += 1.0
                break

    
    return Dic
```

```python
# Lever rule stuff

def leverRule(c_0, c_a, c_b):
    """Calculates phase fractions"""
    Vfrac = (c_0 - c_a)/(c_b - c_a) 
    return Vfrac
```

```python
Al4_5Cu_Comps = {'c_0': 4.5, 'c_a': 0.5, 'c_b': 53}
Al15Cu_Comps = {'c_0': 15, 'c_a': 0.5, 'c_e': 34}
Mg_9Al_Comps = {'c_0': 9, 'c_a': 3, 'c_b': 42}

Al4_5CuVFrac = leverRule(Al4_5Cu_Comps['c_0'], Al4_5Cu_Comps['c_a'], Al4_5Cu_Comps['c_b'])
Al15CuVFrac = leverRule(Al15Cu_Comps['c_0'], Al15Cu_Comps['c_a'], Al15Cu_Comps['c_e'])
Mg_9Al_VFrac = leverRule(Mg_9Al_Comps['c_0'], Mg_9Al_Comps['c_a'], Mg_9Al_Comps['c_b'])
```

```python
#Al15Cu superplot
greens = ['g', '#45DD12', '#2EC20A', '#046307']
fig, axs = plt.subplot_mosaic([['A)', 'B)'], ['C)', 'D)']], figsize=(12, 8),
                              layout='constrained')

for label, ax in axs.items():
    # label physical distance in and down:
    trans = mtransforms.ScaledTranslation(-70/72, -5/72, fig.dpi_scale_trans)
    ax.text(0.0, 1.0, label, transform=ax.transAxes + trans,
            fontsize='18', verticalalignment='top',
            bbox=dict(facecolor='w', edgecolor='none', pad=3.0))
    
fontSize = '16'

### A)

grainSizeAvgAl15 = [Al_15Cu_Grains['0h'].mean(), 
                 Al_15Cu_Grains['8h'].mean(), 
                 Al_15Cu_Grains['12h'].mean(), 
                 Al_15Cu_Grains['16h'].mean(), 
                 Al_15Cu_Grains['20h'].mean()]

grainSizeStdAl15 = [Al_15Cu_Grains['0h'].std(), 
                 Al_15Cu_Grains['8h'].std(), 
                 Al_15Cu_Grains['12h'].std(), 
                 Al_15Cu_Grains['16h'].std(), 
                 Al_15Cu_Grains['20h'].std()]

axs['A)'].errorbar(Al_15Cu_Grains['time'][0:5], grainSizeAvgAl15, grainSizeStdAl15, markersize = 7,fmt = '-o', label = '$Grain\:size$', capsize=3, color = 'g')
axs['A)'].set_ylabel('$Grain\:size\: /\:\:µm^{2}$', fontsize = fontSize)
axs['A)'].set_xlabel('Ageing time / h', fontsize = fontSize)

# Do twinax shit with phase fraction // update y2 with numbers from natalie cos these are whack

y2 = (1 - np.array([Al_15Cu_Grains['0h'].sum(), 
                 Al_15Cu_Grains['8h'].sum(), 
                 Al_15Cu_Grains['12h'].sum(), 
                 Al_15Cu_Grains['16h'].sum(), 
                 Al_15Cu_Grains['20h'].sum()])/Al_15Cu_Grains['total_area'][0:5])*100

ax5 = axs['A)'].twinx()
ax5.plot(Al_15Cu_Grains['time'][0:5], y2, '-', marker = 's', label = 'Eutectic fraction', color = '#45DD12')
ax5.tick_params(axis='both', which='major', labelsize=fontSize)

axs['A)'].plot([], [], color = '#45DD12', marker = 's', label = 'Eutectic fraction')
ax5.set_ylabel('Eutectic fraction / %', fontsize = fontSize)

axs['A)'].set_xticks(np.arange(0,24,4), fontsize = fontSize)
axs['A)'].tick_params(axis='both', which='major', labelsize= fontSize)
axs['A)'].legend(title = "$Al-15Cu\:A_{200\:^{o}C}$",  loc = 'upper right', fontsize = '12', title_fontsize = '14')

### B)

axs['B)'].imshow(imAl15Cu0h)
axs['B)'].set_xticks([])
axs['B)'].set_yticks([])
axs['B)'].set_xlabel("24 h solutionised at $530\:^{o}C$", fontsize = fontSize)

### C)

axs['C)'].imshow(imAl15Cu8h)
axs['C)'].set_xticks([])
axs['C)'].set_yticks([])
axs['C)'].set_xlabel("8 h aged at $200\:^{o}C$", fontsize = fontSize)

### D)

axs['D)'].imshow(imAl15Cu20h)
axs['D)'].set_xticks([])
axs['D)'].set_yticks([])
axs['D)'].set_xlabel("20 h aged at $200\:^{o}C$", fontsize = fontSize)

fig.savefig('Al15Cu grain size.jpg', dpi=400)
```
